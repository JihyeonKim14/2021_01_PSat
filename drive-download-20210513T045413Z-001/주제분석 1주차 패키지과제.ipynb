```python
import math
data['sin_time'] = data['transaction_date'].apply(lambda x: math.sin(x*math.pi*2 / 12*3))
data['cos_time'] = data['transaction_date'].apply(lambda x: math.cos(x*math.pi*2 / 12*3))

data = data.drop(columns = ['year_of_completion','transaction_year_month','transaction_date'])
data = data.drop(columns = ['transaction_month','transaction_month_date'])
```

```python
# 삼각변환을 통한 장점.
# 제가 이해한 바로는 00:02와 23:58은 시간적으로는 4분 차이이지만, 코딩을 할때 인덱싱으로
# 1000이 넘는 차이가 나기 때문에 머신러닝에서 머신에게 시간적 개념을 알려주기 어렵다는 것입니다.
# 이 삼각변환을 통해서 삼각함수들은 주기성을 띄니 그 시간의 차이를 머신에게 진짜 시간적 개념으로 입력할 수 있다라고 이해했습니다.

```

```python
import re
apt=[]
for i in data.apt:
    hangul=re.compile('[^ ㄱ-ㅣ가-힣]+') # 한글과 띄어쓰기를 제외한 모든 글자(정규표현식)
    repl=''
    result=re.sub(pattern=hangul, repl=repl, string=i)
    apt.append(result)    
data.apt=apt
data.head()
 # https://jokergt.tistory.com/52 참고링크
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vec=CountVectorizer()
m=count_vec.fit_transform(apt)
m.toarray()
# https://wiserloner.tistory.com/917 참고링크
# https://programmers.co.kr/learn/courses/21/lessons/1695 참고링크
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()
tfidf_vectorizer.fit(apt)
tfidf_vectorizer.vocabulary_
sorted(tfidf_vectorizer.vocabulary_.items())
tfidf_vectorizer.idf_
tfidf_vectorizer.transform(apt).toarray()
# https://chan-lab.tistory.com/24 참고 링크
```
